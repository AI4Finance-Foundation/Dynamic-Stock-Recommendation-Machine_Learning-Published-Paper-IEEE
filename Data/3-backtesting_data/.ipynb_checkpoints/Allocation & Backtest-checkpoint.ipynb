```python
def rand_weights(n):
    ''' Produces n random weights that sum to 1 '''
    k = np.random.rand(n)
    return k / sum(k)

def equal_portfolio(pred_y, returns):
    ''' 
    Returns the mean and standard deviation of returns for a random portfolio
    '''

    p = np.asmatrix(pred_y)
    #w = np.asmatrix(rand_weights(len(pred_y))) # random weighted
    w = np.asmatrix([1/21]*len(pred_y)) # equally weighted
    C = np.asmatrix(np.cov(returns.T))
    
    mu = w * p.T
    sigma = np.sqrt(w * C * w.T)
    
    # This recursion reduces outliers to keep plots pretty
    if sigma > 2:
        return equal_portfolio(returns)
    
    mu = mu.tolist()
    sigma = sigma.tolist()
    return mu[0][0], sigma[0][0] 

mean = []
stds = []
means_tmp, stds_tmp = random_portfolio(select['predicted_return'], return_rolling)
```

### Markowitz optimization and the Efficient Frontier

```python
def optimal_portfolio(pred_y, returns):
    n = len(returns)
    returns = np.asmatrix(returns)
    n_asset = returns.shape[1]
    
    N = 100
    mus = [10**(5.0 * t/N - 1.0) for t in range(N)]
    
    # Convert to cvxopt matrices
    S = opt.matrix(np.cov(returns.T))
    #pbar = opt.matrix(np.mean(returns, axis=1))
    pbar = opt.matrix(pred_y)
    
    # Create constraint matrices
    G1 = -np.eye(n_asset)
    G2 = np.eye(n_asset)
    G = opt.matrix(np.vstack((G1, G2)))
    h1 = [0]*n_asset
    h2 = [0.1]*n_asset
    h = opt.matrix(np.hstack((h1, h2)))
    A = opt.matrix(1.0, (1, n_asset))
    b = opt.matrix(1.0)
    
    # Calculate efficient frontier weights using quadratic programming
    portfolios = [solvers.qp(mu*S, -pbar, G, h, A, b)['x'] 
                  for mu in mus]
    ## CALCULATE RISKS AND RETURNS FOR FRONTIER
    returns = [blas.dot(pbar, x) for x in portfolios]
    risks = [np.sqrt(blas.dot(x, S*x)) for x in portfolios]
    ## CALCULATE THE 2ND DEGREE POLYNOMIAL OF THE FRONTIER CURVE
    m1 = np.polyfit(returns, risks, 2)
    x1 = np.sqrt(m1[2] / m1[0])
    # CALCULATE THE OPTIMAL PORTFOLIO
    wt = solvers.qp(opt.matrix(x1 * S), -pbar, G, h, A, b)['x']
    return np.asarray(wt), returns, risks

weights, returns, risks = optimal_portfolio(select['predicted_return'], return_rolling)
```

```python
#P&L return: equally
portfolio_3 = portfolio_equal.loc[0,].copy()
portfolio_3.index = range(len(portfolio_3))
tmp = portfolio_3.shift(1).copy()
tmp[0] = 0
tmp.index = portfolio.columns
portfolio_equal.columns = portfolio.columns
portfolio_3 = portfolio_equal.loc[0,] - tmp
for i in range(len(portfolio_3)):
    if(tmp[i] != 0):
        portfolio_3[i] = portfolio_3[i]/tmp[i]

portfolio_3.index = portfolio_equal.columns
portfolio_3['1995-06-01'] = 0
# accumulative
initial = 0
for i in range(portfolio_3.shape[0]):
    portfolio_3[i] = initial + portfolio_3[i]
    initial = portfolio_3[i]
```
